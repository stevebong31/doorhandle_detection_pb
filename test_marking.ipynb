```python
#!/usr/bin/env python  
import roslib
import rospy
import math
import tf
import geometry_msgs
import visualization_msgs.msg 

def make_marker(marker_name,trans,rot,color):
    marker.header.frame_id = 'odom'
    marker.header.seq = 1
    marker.header.stamp = rospy.Time.now()
    marker.type = visualization_msgs.msg.Marker.TEXT_VIEW_FACING
    marker.ADD
    marker.text = marker_name

    marker.pose.position.x = trans[0]
    marker.pose.position.y = trans[1]
    marker.pose.position.z = trans[2]
    marker.ns = marker_name
    marker.pose.orientation.x =rot[0] 
    marker.pose.orientation.y =rot[1] 
    marker.pose.orientation.z =rot[2] 
    marker.pose.orientation.w =rot[3] 

    marker.scale.x = 0.3
    marker.scale.y = 0.3
    marker.scale.z = 0.3
    marker.color.a = 1
    if color == 0:
        marker.color.r = 0
        marker.color.g = 1
        marker.color.b = 0
    elif color == 1:
        marker.color.r = 1
        marker.color.g = 0
        marker.color.b = 0
    else:
        marker.color.r = 0
        marker.color.g = 0
        marker.color.b = 1

    markerarr.markers.append(marker)

rospy.init_node('pepper_tf2move')

listener = tf.TransformListener()

pose_stamp =geometry_msgs.msg.PoseStamped()

door_marker = rospy.Publisher('/marker', visualization_msgs.msg.MarkerArray,queue_size=1)


marker = visualization_msgs.msg.Marker()
markerarr = visualization_msgs.msg.MarkerArray()


tmp_m = 0
tmp_o1 = 0
tmp_o2 = 0
c = 8
while True:
    try:
        (trans1,rot1) = listener.lookupTransform('/odom','/meeting_room_0',rospy.Time(0))
        #print(trans1,rot1)
        tmp_m += 1 
        if tmp_m == c:
            make_marker('meeting_room',trans1,rot1,0)

            door_marker.publish(markerarr)
    except:
        tmp_m = 0

    try:
        (trans2,rot2) = listener.lookupTransform('/odom','/office_room1_0',rospy.Time(0))
        #print(trans2,rot2)
        tmp_o1 += 1
        if tmp_o1 == c:
            make_marker('room2',trans2,rot2,1)
            door_marker.publish(markerarr)
    except:
        tmp_o1 = 0

    try:
        (trans3,rot3) = listener.lookupTransform('/odom','/office_room2_0',rospy.Time(0))
        #print(trans3,rot3)
        tmp_o2 += 1
        if tmp_o2 == c:
            make_marker('room3',trans3,rot3,2)

            door_marker.publish(markerarr)
    except:
        tmp_o2 = 0
```
